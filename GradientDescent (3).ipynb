```python
class LinearRegression:
    def __init__(self, w0,w1,max_iteration,lr=0.001, metric_change = 0.0000000001):
        self.lr = lr
        self.w0 = w0
        self.w1 = w1
        self.max_iteration = max_iteration
        self.metric_change = metric_change
    
    def cost_function(self,x,y):
        n = len(x)
       
        
        pDw0 = (2/n) * np.sum((self.w0 + self.w1*x) - y)
        pDw1 = (2/n) * np.sum((self.w0 + self.w1*x - y) * x)
        
        return pDw0, pDw1
        
        
    def update_weights(self, x,y):
        
        pDw0, pDw1 = self.cost_function(x,y)
        
        self.w0 = self.w0 - self.lr * pDw0
        self.w1 = self.w1 -self.lr * pDw1
        
        print(self.w0,self.w1)
        
    #def checkMetricChange(self,L):
        
    
    def fit(self, x_train, y_train):
        for i in range(self.max_iteration):
            
            L =  np.mean((self.w0 + self.w1*x_train - y_train)**2)
            
            self.update_weights(x_train,y_train)
            
            L_change = L -  np.mean((self.w0 + self.w1*x_train - y_train)**2)
            
            if (L_change <= self.metric_change):
   
                 break
            
            
            
        
    def predict(self, x_test):
        
        y_pred = self.w0 + self.w1*x_test
        return y_pred
```

```python
class RidgeRegression:
    def __init__(self, max_iteration, lamb, lr=0.0001, metric_change = 0.0000000000000001 ):
        self.lr = lr
        self.max_iteration = max_iteration
        self.metric_change = metric_change
        self.lamb = lamb
        
    
    def cost_function(self,x,y):
        m = len(y)
        z = x.dot(self.w) + self.w0 - y  
        
        pDloss_w = (1/m) *(x.T).dot(z) + 2*self.lamb* np.sum(self.w)
        pDloss_w0 = (1/m) *np.sum(z)
        
        return pDloss_w0, pDloss_w
    
    def update_weights(self,x,y):
        
        pDloss = self.cost_function(x,y)
        
        self.w0 = self.w0 - self.lr*pDloss[0]
        self.w = self.w-self.lr*pDloss[1]
         
    
    def fit(self, x_train, y_train):
        self.w = np.random.rand(x_train.shape[1])
        self.w0 = np.random.rand()
        for i in range(self.max_iteration):
            
            
            Loss = (1/2)*np.mean((x_train.dot(self.w) + self.w0 -y_train)**2)+ np.sum(self.w**2)
            self.update_weights(x_train,y_train)
            L_change = Loss - (1/2)*np.mean((x_train.dot(self.w) + self.w0 -y_train)**2) + np.sum(self.w**2)
            if L_change <= self.metric_change:
                break
        return self.w0, self.w   
        
    def predict(self, x_test):
        
        y_predict = (x_test).dot(self.w)+ self.w0
        return y_predict
```
